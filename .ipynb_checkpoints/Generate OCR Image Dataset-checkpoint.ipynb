```python
all_files = glob.glob(path + "/*.xlsx")
li = []
for filename in all_files:
    print(filename)
    df = pd.read_excel(filename, index_col=None, header=0)
    li.append(df)
```

```python
df = pd.concat(li, axis=0, ignore_index=True)
```

```python
df.tail()
```

```python
dictionary_gum = set()
df['gum_text'].str.lower().str.split().apply(dictionary_gum.update)
dictionary_gum = sorted(dictionary_gum)
print(dictionary_gum[:5]) # limit log to 5 items

# TODO: Augmentation. add numbers randomly

# Saves dictionary for further reference
filename = "dictionary_gum"
version = "1" # change on every run

with open(f'{filename}_{version}.txt', 'w', encoding="utf-8") as f:
    for line in dictionary_gum:
        f.write(line)
        f.write('\n')
```

```python
sentences_gum = df['gum_text'].to_list() #dictionary_gum #TODO: change
print(sentences_gum[:5])
print(len(sentences_gum))
```

```python
# Split sentences by punctuation
import re
import numpy as np
import string

def split_sentences(sentences):
    result = []
    for sentence in sentences:
        # split the string using the pattern
        temp = re.findall('.*?[.;,]', sentence)
        res = [f"{ele.strip()}" # 3% chance of having a dot
                .strip() for ele in temp if ele] 
        result.extend(res)
    return result

sentences_gum = split_sentences(sentences_gum)
print(sentences_gum[-6:])
print(len(sentences_gum))
```

```python
# Adds sentences list with single unique words (dictionary)
sentences_gum.extend(dictionary_gum)
```

```python
sentences_gum.extend([*string.digits])
```

```python
print(sentences_gum[:5]) # First 5 items
print(sentences_gum[-5:]) # Last 5 items
print(len(sentences_gum))
```

```python
trans_dict = {
             #ord('-'): '',
             ord('–'): '-',
             ord('—'): '-',
             #ord('“'): '"',
             #ord('”'): '"',
            #  ord('.'): '.',
            #  ord('!'): '',
            #  ord('¡'): '',
            #  ord('?'): '',
            #  ord('¿'): '',
             #ord('’'): '"',
             #ord('‘'): '"',
             #ord("'"): '"',
             ord('|'): '',
             #ord('#'): '', 
             #ord('$'): '',
             #ord('%'): '', 
             ord('&'): '',
            #  ord(';'): ',',
            #  ord('"'): '',
             #ord('«'): '"',|
             #ord('»'): '"',
             ord('ø'): 'ɵ', # Ref. https://en.wikipedia.org/wiki/Close-mid_central_rounded_vowel 
             ord('Ø'): 'Ɵ', # Mayus
        }
for idx, sentence in enumerate(sentences_gum):
    sentences_gum[idx] = str(sentence).strip().translate(trans_dict)
```

```python
# Remove empty strings
sentences_gum = list(filter(None, sentences_gum))
```

```python
# The generators use the same arguments as the CLI, only as parameters
import os
import math
from random import choice

total_train_ratio = 0.8
image_total_per_gen = 100

custom_fonts = [
            "fonts/ARIAL.TTF",
            "fonts/ARIALBD.TTF",
            "fonts/ARIALBI.TTF",
            "fonts/ARIALI.TTF",
            "fonts/CALIBRI.TTF",
            "fonts/CALIBRIB.TTF",
            "fonts/CALIBRII.TTF",
            "fonts/CALIBRIL.TTF",
            "fonts/CALIBRILI.TTF",
            "fonts/CALIBRIZ.TTF",
            "fonts/MICROSS.TTF",
            "fonts/TAHOMA.TTF",
            "fonts/TAHOMABD.TTF",
            "fonts/TIMES.TTF",
            "fonts/TIMESBD.TTF",
            "fonts/TIMESBI.TTF",
            "fonts/TIMESI.TTF",
            #os.path.join(os.path.dirname(os.path.abspath("__file__")), "fonts", font)
            #   for font in os.listdir(
            #      os.path.join(os.path.dirname(os.path.abspath("__file__")), "fonts")
            #  )
        ]

# Black text and white background
generator_wbgbt = GeneratorFromStrings(
    sentences_gum,
    fonts=custom_fonts,
    size = 50,
    blur=1,
    random_blur=True,
    count=image_total_per_gen,  # otherwise -1, all
    skewing_angle=1,
    random_skew=True,
    background_type=1, # Gaussian noise white bg
    distorsion_type=2,
)

# Custom background generator with black text
generator_cbgbt = GeneratorFromStrings(
    sentences_gum, 
    fonts=custom_fonts,
    size = 40,
    blur=1,
    random_blur=True,
    count=image_total_per_gen,  # otherwise -1, all
    skewing_angle=3,
    random_skew=True,
    background_type=3,
    image_dir="./TextRecognitionDataGenerator/trdg/images",
)

# Custom background generator with white text
generator_cbgwt = GeneratorFromStrings(
    sentences_gum,
    fonts=custom_fonts,
    size = 40,
    blur=1,
    random_blur=True,
    count=image_total_per_gen,  # otherwise -1, all
    skewing_angle=3,
    random_skew=True,
    background_type=3,
    text_color = "#ffffff",
    image_dir="./TextRecognitionDataGenerator/trdg/images",
)

# Saving configuration
generated = []
count = 0
generated_path = 'trainer/all_data/gum_train'
val_path = 'trainer/all_data/gum_val'

os.makedirs(generated_path, exist_ok=True)
os.makedirs(val_path, exist_ok=True)

generators = [generator_wbgbt, generator_cbgbt, generator_cbgwt] # Order defines saved type

for idx, generator in enumerate(generators):
    for img, lbl in generator:
        if img:
            img.save(f'{generated_path}/{count}.jpg') # .show()
            generated.append([f'{count}.jpg', lbl, idx])
            count += 1

# Create the pandas DataFrame
df_generated = pd.DataFrame(generated, columns=['filename', 'words', 'type'])

# Shuffle dataframe using sample function
df = df_generated.sample(frac=1)

# Define train / validation sizes
train_size = math.floor(count*total_train_ratio)
val_size = count - train_size

# Split data into test and train
train = df[0:train_size]
val = df[train_size:]

# print train set
print("Train dataframe")
print(train)

# print test set
print("Validation dataframe")
print(val)
```

```python
import shutil

files_to_move = val["filename"].values.tolist()

# iterate files
for file in files_to_move:
    # construct full file path
    source = generated_path + '/' + file
    destination = val_path + '/' + file
    # move file
    shutil.move(source, destination)
    #print('Moved:', file)
```

```python
# Save in CSV with all data
train.to_csv(f'{generated_path}/labels_all.csv', index=False, encoding='utf-8')
val.to_csv(f'{val_path}/labels_all.csv', index=False, encoding='utf-8')

# Save in CSV for trainer
train.to_csv(f'{generated_path}/labels.csv', columns=['filename', 'words'], index=False, encoding='utf-8')
val.to_csv(f'{val_path}/labels.csv', columns=['filename', 'words'], index=False, encoding='utf-8')
```
